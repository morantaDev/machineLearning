```python
path = '/home/moranta/Downloads/données_ventes-20231006T170508Z-001/données_ventes/'
csv_files = glob.glob(path + "/*.csv")
csv_files
```

```python
df_list = (pd.read_csv(file) for file in csv_files)
df_list
```

```python
data = pd.concat(df_list, ignore_index=True)
data
```

```python
data.to_csv("data.csv")
data_copy = data.copy()
data_copy
```

```python
data = data.dropna()
data = data.drop_duplicates()
data.isnull().sum()
# data['Order Date'] = pd.to_datetime(data['Order Date'])
```

```python
# Convertir la colonne de date en un format de date si ce n'est pas déjà fait
data = data[data['Order Date'] != 'Order Date']
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%m/%d/%y %H:%M')
data.isnull().sum()
```

```python
# Extraire le mois de chaque date
data['mois'] = data['Order Date'].dt.month
data
```

```python
data['Price Each'] = pd.to_numeric(data['Price Each'], errors='coerce')
# Regrouper par mois et calculer le chiffre d'affaires total
chiffre_affaires_mois = data.groupby('mois')['Price Each'].sum()
chiffre_affaires_mois

import matplotlib.pyplot as plt

# Créer un graphique
plt.figure(figsize=(10, 6))
chiffre_affaires_mois.plot(kind='bar')
plt.xlabel('Mois')
plt.ylabel('Chiffre d\'affaires')
plt.title('Chiffre d\'affaires mensuel')
plt.grid(True)
plt.show()
```

```python
# Trouver le mois avec le chiffre d'affaires le plus élevé
mois_meilleur_chiffre_affaires = chiffre_affaires_mois.idxmax()
mois_meilleur_chiffre_affaires
```

```python
# Extraire la ville à partir de la colonne "Purchase Address"
data['Ville'] = data['Purchase Address'].apply(lambda x: x.split(',')[1])

# Compter le nombre de commandes par ville
commandes_par_ville = data['Ville'].value_counts()

# Créer un graphique à barres horizontales
plt.figure(figsize=(10, 6))
commandes_par_ville.plot(kind='barh')
plt.xlabel('Nombre de commandes')
plt.ylabel('Ville')
plt.title('Nombre de commandes par ville')
plt.grid(True)
plt.show()
```

```python
# Compter le nombre de commandes par ville
ville_max_commandes = data['Ville'].value_counts().idxmax()
ville_max_commandes
```

```python
data['Quantity Ordered'] = pd.to_numeric(data['Quantity Ordered'], errors='coerce')
# Compter le nombre de ventes par produit
produits_plus_vendus = data.groupby('Product')['Quantity Ordered'].sum().sort_values(ascending=False)
produits_plus_vendus
```

```python
# Créer un graphique des produits les plus vendus
plt.figure(figsize=(12, 6))
produits_plus_vendus.head(10).plot(kind='bar')
plt.xlabel('Produit')
plt.ylabel('Quantité vendue')
plt.title('Les produits les plus vendus')
plt.grid(True)
plt.show()
```

```python
# Créer une nouvelle colonne avec les combinaisons de produits
data['Combinaison de Produits'] = data.groupby('Order ID')['Product'].transform(lambda x: ', '.join(x))

# Compter les combinaisons de produits les plus vendues
combinaisons_plus_vendues = data['Combinaison de Produits'].value_counts().head(10)


# Créer un graphique des combinaisons de produits les plus vendues
plt.figure(figsize=(12, 6))
combinaisons_plus_vendues.plot(kind='bar')
plt.xlabel('Combinaison de Produits')
plt.ylabel('Nombre de ventes')
plt.title('Combinaisons de produits les plus vendues')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```
